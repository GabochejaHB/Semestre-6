```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
plt.style.use("bmh")
```

```python
def recta(x,A,B):
    return x*A + B
ajuste_recta= curve_fit(recta, angulos, altura)[0]
```

```python
altura = np.array([41.02,43.55,44])
angulos = np.array([35,40,45])

xx = np.linspace(35,45,10**3)
def graf(A=1,B=1):
    plt.scatter(angulos, altura)
    plt.plot(xx, recta(xx, A,B), color="red")
    plt.xlabel(r"$\theta [Â°]$")
    plt.ylabel(r"Altura [m]")
    plt.grid(True)
    chi2= np.sum((altura - recta(angulos,A,B))**2)
    print(chi2)
```

```python
graf(ajuste_recta[0], ajuste_recta[1])
```

```python
res = altura - recta(angulos, ajuste_recta[0], ajuste_recta[1])
st = np.sum((altura - np.mean(altura))**2)
sr = np.sum(res**2)
r2 = (st - sr)/st
print(r2)
```
