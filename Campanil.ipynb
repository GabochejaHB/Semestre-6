```python
#altura = np.array([41.41, 41.86,40.23,40.72, 42.09, 41.56])
#angulos = np.array([0.925,0.855,0.768,0.715,0.681,0.628])
#--------------------
#altura = np.array([41.41, 41.86,41.57,42.11, 43.5, 41.56])
#angulos = np.array([0.925,0.855,0.785,0.733,0.698,0.628])
#--------------------
altura = np.array([41.57,42.11,42.08,41.56])
angulos = np.array([0.785,0.733,0.681,0.628])

L = np.array([40,45,50,55])

# ERROR ANALITICO
error_angulos = 0.0087
error_altura = np.sqrt(0.0005**2 + (np.tan(angulos)*0.0005)**2 + (L*(1/np.cos(angulos)**2)*0.0087)**2)
print(np.mean(altura), np.std(altura, ddof = 1))
print(error_altura)
print(np.std(altura, ddof = 1))
```

```python
def recta(x,A,B):
    return x*A + B
ajuste_recta = curve_fit(recta, angulos, altura)[0]
```
