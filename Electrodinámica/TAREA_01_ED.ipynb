#  **d)** Gráfico del potencial eléctrico

En este ítem se utilizó el resultado calculado en **(a)** para gráficar el potencial respecto a la distancia al centro de la bola. Esto es:

$$ \Phi(z)=
  \left \{
    \begin{aligned}
      \frac{kQ}{2R^3}\left( 3R^2 - z^2 \right)&; z \leq R \\
      \frac{kQ}{z} &; z \geq R.
    \end{aligned}
  \right .
$$

En el gŕafico se consideró el radio de la bola $R=1$[m] y la carga $Q=1$[C]. Además, suponemos que trabajamos con una distancia $z$ medida en metros. De esta forma, consideramos el potencial medido en Voltios [V].

```python
e_0 = 8.85*10**(-12) #Epsilon.
k = 1/(4*np.pi*e_0) # Constante eléctrica.

# Definimos la función Potencial dentro de la bola.
def potencial_in(r, R=1, Q=1):
    return (k*Q*0.5)/(R**3)*(3*R**2 - r**2) 

#Definimos la función Potencial fuera de la bola.
def potencial_out(r,Q=1):
    return k*Q/r
```

```python
#Gráfico de la función Potencial
r1= np.linspace(0,1,10**3)
r2 = np.linspace(1,5,10**3)
plt.plot(r1,potencial_in(r1), label = r"$\Phi(r) \longrightarrow r \leq R$")
plt.plot(r2,potencial_out(r2), label = r"$\Phi(r) \longrightarrow r \geq R$")
plt.xlabel(r"$r$")
plt.ylabel(r"$ \Phi(r)$")
plt.title("$\Phi(r)$")
plt.legend()
plt.savefig("Potencial.pdf")
plt.show()
```

# **e)** Gráfico del Campo Eléctrico

En este ítem se utilizó el resultado calculado en **(b)** para gráficar el modulo del campo eléctrico respecto a la distancia al centro de la bola. Esto es:

$$ E(z)=
  \left \{
    \begin{aligned}
      \frac{kQz}{R^3}&; z \leq R \\
      \frac{kQ}{z^2} &; z \geq R.
    \end{aligned}
  \right .
$$

En el gŕafico se consideró el radio de la bola $R=1$[m] y la carga $Q=1$[C]. Además, suponemos que trabajamos con una distancia $z$ medida en metros. De esta forma, consideramos el Campo Eléctrico medido en voltios por metro [V/m].

```python
#Definimos el Campo Eléctrico dentro de la bola.
def camp_in(r,Q=1,R=1):
    return k*Q*r/R**3

#Definimos el Campo Eléctrico fuera de la bola.
def camp_out(r,Q=1):
    return k*Q/r**2
```

```python
#Gráfico del Campo Eléctrico.
plt.plot(r1,camp_in(r1), label = r"$E \longrightarrow r \leq R$")
plt.plot(r2,camp_out(r2), label = r"$E \longrightarrow r \geq R$")
plt.xlabel(r"$r$")
plt.ylabel(r"$E$")
plt.legend()
plt.title("$E(r)$")
plt.savefig("Campo_electrico.pdf")
plt.show()
```

# **g)** Gráfico del Campo Eléctrico generado por dos bolas de carga opuesta

En este ítem se utilizó el resultado calculado en **(g)** para gráficar el Campo Eléctrico generado por dos bolas de carga opuesta separadas una distancia $d$ desde sus centros. Esto es:

$$ \vec{E}(\vec{x}) = kQ \left[ \frac{x\hat{x}+y\hat{y}+z\hat{z}}{\left( x^2+ y^2 + z^2 \right)^{3/2}}  - \frac{(x-d)\hat{x}+y\hat{y}+z\hat{z}}{\left( (x-d)^2+ y^2 + z^2 \right)^{3/2}}\right]. $$

En el gŕafico se consideró la distancia entre las bolas $d=3$[m] y la carga $Q=1$[C]. Además, suponemos que trabajamos con las distancias $x,y,z$ medidas en metros. De esta forma, consideramos el Campo Eléctrico medido en Voltios por metro [V/m]].

```python
#Definimos la malla a usar
n = 20 # nodos de la grilla
max = 5# máximo y mínimo del gráfico
X = np.linspace(-max,max,n)
Y = np.linspace(-max,max,n)
x,y = np.meshgrid(X,Y)
phi = np.linspace(0,2*np.pi, n)
```

```python
# Definimos la componente del Campo Eléctrico en x
def Ex(x,y,d=3,Q=1,R=1): 
    b = x-d
    r1 = (x**2 + y**2)**1.5
    r2 = (b**2 + y**2)**1.5
    return k*Q*(x/r1 - b/r2)

# Definimos la componente del Campo Eléctrico en y
def Ey(x,y,d=3,Q=1,R=1):
    b = x-d
    r1 = (x**2 + y**2)**1.5
    r2 = (b**2 + y**2)**1.5
    return k*Q*(y/r1 - y/r2)

# Definimos la magnitud del Campo Eléctrico para posterior normalización.
def E(x,y):
    E = np.sqrt(Ex(x,y)**2 + Ey(x,y)**2) 
    return E
```

```python
#Graficamos el Campo Eléctrico Normalizado.
plt.figure(figsize=(8,5))
plt.quiver(x,y,Ex(x,y)/E(x,y),Ey(x,y)/E(x,y), E(x,y), cmap="seismic", color='blue', scale=50, pivot='middle')
plt.xlim(-1.1*max,1.1*max)
plt.ylim(-1.1*max,1.1*max)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.colorbar()
plt.title('$\\vec{E}(\\vec{x})$')
plt.savefig("Campo_electrico_malla.pdf")
```
