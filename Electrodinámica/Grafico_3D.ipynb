```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.7
phi = 2*np.pi*0.5

def flux_qubit_potential(x,y):
    return 2 + alpha - 2 * np.cos(2*np.pi*x)*np.cos(2*np.pi*y) - alpha*np.cos(phi-2*x)
```

```python
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X,Y = np.meshgrid(x, y)
Z = flux_qubit_potential(X, Y).T
```

```python
ax = plt.subplot(1, 1, 1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0, antialiased=False)
```

ax = plt.subplot(1, 1, 1, projection='3d')
ax.plot_wireframe(X, Y, Z, rstride=4, cstride=4)

```python
# Cambio de Perspectiva

plt.figure(figsize=(12,6))

ax=plt.subplot(1,2,1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
ax.view_init(30, 45) #Cambio de Orientacion

ax=plt.subplot(1,2,2, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
ax.view_init(70, 30) # Cambio de Orientacion
```

# La API orientada al objeto

```python
fig = plt.figure()

ejes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # izquierda, abajo, ancho, altura (rango 0 a 1)

ejes.plot(x,y, color = 'r')

ejes.set_xlabel('x')
ejes.set_ylabel('y')
ejes.set_title(u'Ttulo');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # ejes principales
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # ejes del grfico insertado

# figura principal
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title(u'Ttulo principal')

# figura insertada
axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title(u'Ttulo secundario')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

#Primer grfico
axes[0].plot(x, y, 'r')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
plt.tight_layout()
axes[0].set_title(u'Ttulo')

#Segundo Grfico
axes[1].plot(y,x, 'r')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
plt.tight_layout()
axes[1].set_title(u'Ttulo')
```
