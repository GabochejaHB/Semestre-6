```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
plt.style.use("bmh")
```

```python
# PRIMER METODO
radio = np.array([4,4.5,5,8]) # En cm
perimetro = np.array([25.2,28.2,31.4,50.2]) #en cm

err = 0.05 # en cm
pi_1 = perimetro/(2*radio)
error_pi_1 = np.sqrt((1/(2*radio) * err)**2 + (perimetro/(2*radio**2) * err)**2)
```

```python
plt.errorbar(radio, pi_1, xerr= err, yerr=error_pi_1, fmt = "o", color = "darkblue",ecolor = "darkblue")
plt.hlines(np.mean(pi_1),4,8, ls = "--", color = "magenta", label=r"Promedio de $\pi$.")
plt.hlines(np.pi, 4,8, label = r"Valor real de $\pi$")
plt.ylabel(r"$\pi$")
plt.xlabel(r"$Radio [cm]$")
plt.legend()
plt.savefig("pi_metodo1.pdf")
```

```python
print(pi_1, error_pi_1, np.std(pi_1), np.mean(pi_1))
```

```python
# SEGUNDO METODO
Radio = np.array([4,4.5,5,8]) #en centímetros
circulo = np.array([1.02603, 1.50946, 6.07361, 14.97495]) #en gramos
cuadrado = np.array([0.33134, 0.48270,1.88232, 4.75779])#en gramos

pi_2 = circulo/cuadrado

err_gramo = 10**(-5) #en gramos
err_pi_2 = np.sqrt(((1/cuadrado) * err_gramo)**2 + ((circulo/(cuadrado)**2) * err_gramo)**2)
```

```python
print(pi_2, err_pi_2, np.mean(pi_2))
```

```python
plt.errorbar(Radio, pi_2, xerr= err, yerr=err_pi_2, fmt = "o", color = "darkblue",ecolor = "darkblue")
plt.hlines(np.mean(pi_2),4,8, ls = "--", color = "magenta", label=r"Promedio de $\pi$.")
plt.hlines(np.pi, 4,8, label = r"Valor real de $\pi$")
plt.ylabel(r"$\pi$")
plt.xlabel(r"$Radio [cm]$")
plt.legend()
plt.savefig("pi_metodo2.pdf")
```
